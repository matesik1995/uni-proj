```python
from sympy.interactive import printing
printing.init_printing(use_latex=True)
from __future__ import division
from sympy import *
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = Symbol("x")
eq = x**3+3*x**2-10*x+24
arr = np.arange(-5,5,0.1)
val = [eq.evalf(subs = { x : i }) for i in arr]
```

```python
plt.plot(arr,val)
plt.show()
solve(eq,x)
```

#### 2

```python
x, y = symbols("x y")
eqs = [x**2+3*y-10, 4*x-y**2+2]
sol = solve(eqs, [x, y], dict=True, check=False)
sol
```

```python
for s in sol:
    for k, v in s.items():
        print(k, v.evalf())
```
